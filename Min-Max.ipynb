```python
import networkx as nx
import matplotlib.pyplot as plt

# ------------------ Game Tree ------------------
G = nx.DiGraph()

edges = [
    ('A', 'B'), ('A', 'C'),
    ('B', 'D'), ('B', 'E'),
    ('C', 'F'), ('C', 'G')
]

G.add_edges_from(edges)

# Leaf node values
values = {
    'D': 3,
    'E': 5,
    'F': 2,
    'G': 9
}


# ------------------ Minimax Algorithm ------------------
def minimax(node, is_max_turn):
    # Base case: leaf node
    if node in values:
        return values[node]

    # MAX player
    if is_max_turn:
        best = float('-inf')
        for child in G[node]:
            best = max(best, minimax(child, False))
        return best

    # MIN player
    else:
        best = float('inf')
        for child in G[node]:
            best = min(best, minimax(child, True))
        return best


# ------------------ Run Minimax ------------------
result = minimax('A', True)
print("Optimal Value (Minimax):", result)


# ------------------ Visualization ------------------
pos = {
    'A': (0, 3),
    'B': (-2, 2),
    'C': (2, 2),
    'D': (-3, 1),
    'E': (-1, 1),
    'F': (1, 1),
    'G': (3, 1)
}

labels = {
    node: f"{node}\n({values[node]})" if node in values else node
    for node in G.nodes()
}

plt.figure(figsize=(6, 4))
nx.draw(
    G,
    pos,
    labels=labels,
    node_size=2000,
    node_color='lightgreen',
    edge_color='gray',
    with_labels=True
)

plt.title("Minimax Game Tree")
plt.show()
```
