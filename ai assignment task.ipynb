```python
# Activity 1: Basic Recursive DFS
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = set() # Set to keep track of visited nodes [cite: 46]

def dfs(visited, graph, node): # Function for dfs [cite: 47]
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

# Driver Code
print("Following is the Depth-First Search (Activity 1):")
dfs(visited, graph, '5') # Starting from root node 5 [cite: 54]
```

```python
# Activity 2: DFS for Complete Graphs
from collections import defaultdict

class Graph:
    def __init__(self):
        # Default dictionary to store graph [cite: 76]
        self.graph = defaultdict(list)

    def addEdge(self, u, v):
        self.graph[u].append(v)

    def DFSUtil(self, v, visited):
        # Mark the current node as visited and print it [cite: 81]
        visited[v] = True
        print(v, end=' ')
        # Recur for all the vertices adjacent to this vertex [cite: 85]
        for i in self.graph[v]:
            if visited[i] == False:
                self.DFSUtil(i, visited)

    def DFS(self):
        V = len(self.graph)
        # Mark all the vertices as not visited [cite: 94]
        visited = [False] * (max(self.graph.keys()) + 1)
        # Call the recursive helper function to print DFS traversal [cite: 95]
        for i in range(len(visited)):
            if visited[i] == False:
                self.DFSUtil(i, visited)

# Driver code
g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("Following is Depth First Traversal (Activity 2):")
g.DFS()
print() # For new line
```

```python
# Lab Task 1: DFS_REG.py
def dfs_tsp(graph, start, goal):
    # Stack stores (current_node, path_taken, cumulative_distance)
    stack = [(start, [start], 0)]
    visited = set()

    while stack:
        (node, path, distance) = stack.pop()

        if node == goal:
            return path, distance

        if node not in visited:
            visited.add(node)
            # Add neighbors to stack (reversed to process in a specific order)
            if node in graph:
                for neighbor, weight in reversed(graph[node]):
                    if neighbor not in visited:
                        stack.append((neighbor, path + [neighbor], distance + weight))
    return None, 0

# Graph implementation based on Figure: Travelling Salesman Problem [cite: 162]
tsp_graph = {
    'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
    'Zerind': [('Oradea', 71)],
    'Oradea': [('Sibiu', 151)],
    'Sibiu': [('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara': [('Lugoj', 111)],
    'Lugoj': [('Mehadia', 70)],
    'Mehadia': [('Drobeta', 75)],
    'Drobeta': [('Craiova', 120)],
    'Craiova': [('Pitesti', 138)],
    'Rimnicu Vilcea': [('Pitesti', 97), ('Craiova', 146)],
    'Fagaras': [('Bucharest', 211)],
    'Pitesti': [('Bucharest', 101)],
    'Bucharest': []
}

# Execution
path, total_dist = dfs_tsp(tsp_graph, 'Arad', 'Bucharest')
print("Lab Task 1 Result:")
print(f"Path: {' -> '.join(path)}")
print(f"Total Distance: {total_dist}")
```
