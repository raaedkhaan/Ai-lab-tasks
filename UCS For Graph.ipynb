```python
import heapq # Used to implement a priority queue (min-heap)

# Graph representation (Adjacency List)
graph = {
    'S': [('A', 3), ('B', 2), ('C', 7)],
    'A': [('D', 3), ('E', 8), ('G', 15)],
    'B': [('G', 20)],
    'C': [('G', 6)],
    'D': [], # Terminal node (no further branches)
    'E': [], # Terminal node (no further branches)
    'G': []  # Terminal node (no further branches)
}

def uniform_cost_search(start, goal):

    # Priority Queue stores tuples of:
    # (total_cost, current_node, path_taken)
    pq = []
    # Push the start node with cost 0 and path containing only start
    heapq.heappush(pq, (0, start, [start]))

    # Set to keep track of visited nodes
    visited = set()

    # Loop until the priority queue is empty
    while pq:
        cost, node, path = heapq.heappop(pq)

        # Goal Test
        if node == goal:
            return path, cost

        if node not in visited:
            visited.add(node)

            for neighbor, weight in graph[node]:
                heapq.heappush(
                    pq,
                    (cost + weight, neighbor, path + [neighbor])
                )

    return None


# Run UCS
path, cost = uniform_cost_search('S', 'G')

# Print the optimal (lowest cost) path and its total cost
print("Optimal Path:", path)
print("Total Cost:", cost)
```
